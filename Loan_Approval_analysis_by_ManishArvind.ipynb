```python
# Importing the required libraries

import numpy as np
import pandas as pd
import scipy as sp
```

```python
# Reading the dataset

df=pd.read_csv('la_dataset.csv')
df.head(10)
```

Displaying the Categorical and Numerical variables

```python
# categorical columns

categorical_var=df.select_dtypes(include='object')
print(categorical_var)

print("There are",len(categorical_var.columns),"categorical columns in the dataset")
```

```python
# numerical columns

numerical_var=df.select_dtypes(include='number')
print(numerical_var)

print("There are",len(numerical_var.columns),"numerical columns in the dataset")
```

Checking column with null values

```python
null_values=df.isna().sum().sort_values(ascending=False)
print(null_values)
```

```python
# replacing NULL values with their espective average values

df.fillna(df.mean(),inplace=True)
df.head(20)
```

Checking the loan amount of an average person based on 'Gender', 'Married', 'Self_Employed'. 
This will give a basic idea of the average loan amount of a person.

```python
pd.pivot_table(data=df,index=['Married','Self_Employed'],columns='Gender',values='LoanAmount',aggfunc='mean')
```

check the percentage of loan approved based on a person's employment type.

```python
df[(df['Loan_Status']=='Y')&(df['Self_Employed']=='Yes')]['Self_Employed'].value_counts()*100/len(df)
```

```python
df[(df['Loan_Status']=='Y')&(df['Self_Employed']=='No')]['Self_Employed'].value_counts()*100/len(df)
```

A government audit is happening real soon! So the company wants to find out those applicants with long loan amount term.
Loan Term more than 25 years

```python
df['Loan_Amount_Term']=df['Loan_Amount_Term'].apply(lambda x:x/12)

print(df[df['Loan_Amount_Term']>25])
```

check the average income of an applicant and the average loan given to a person based on their income.

```python
df.groupby('Loan_Status')[['ApplicantIncome','Credit_History']].mean()
```
